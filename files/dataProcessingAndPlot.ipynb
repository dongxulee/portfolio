### Data Processing Part

```python
import yfinance as yf
import pandas as pd

# This function is used to 
def adjustClosePrice(tickerList, start="2019-10-18", end="2020-03-18"):
    df = pd.DataFrame()
    for ticker in tickerList:
        adjPrice = yf.download(ticker, start, end)[["Adj Close"]]
        adjPrice = adjPrice.rename(columns={"Adj Close": ticker})
        if (df.shape == (0,0)):
            df = adjPrice
        else:
            df = pd.merge(df, adjPrice, how = "outer", left_index=True, right_index=True)
    return df.round(2)

etfTickers = ["SPY", "XLF", "IWM", "EEM", "QQQ", "VWO", "EFA", "FXI", "XLI", "TZA", "XLE", "SDS", "SLV",
              "FAZ", "GDX", "EWJ", "SPXU", "EWZ", "FAS", "TVIX", "USO", "GLD", "TBT", "UNG"]
len(etfTickers)
```

```python
df = adjustClosePrice(etfTickers, start="2019-10-18", end="2020-03-18")
```
